````python
import numpy as np          # Importa NumPy para trabajar con arreglos y funciones numéricas
import pandas as pd         # Importa Pandas para trabajar con datos tabulares (DataFrames)
import matplotlib.pyplot as plt  # Importa pyplot para crear gráficos con Matplotlib

plt.style.use("seaborn-v0_8")  # Aplica el estilo visual "ggplot" a todos los gráficos
plt.rcParams["figure.figsize"] = (8, 4)  # Define el tamaño por defecto de las figuras: 8x4 pulgadas
np.random.seed(42)          # Fija la semilla de aleatoriedad para que los resultados sean reproducibles

print("Matplotlib:", plt.matplotlib.__version__)  # Muestra en consola la versión de Matplotlib instalada
````

## Algunos estilos comunes:

- default — Estilo clásico por defecto.
- classic — Apariencia tradicional de las versiones antiguas de Matplotlib.
- seaborn-v0_8 — Visualización limpia y moderna inspirada en Seaborn.
- Solarize_Light2 — Colores cálidos, contraste suave.
- fivethirtyeight — Estilo periodístico moderno (FiveThirtyEight).
- bmh — Ideal para análisis estadístico.
- grayscale — Gráficos en blanco y negro (útil para impresión).
- dark_background — Fondo negro, útil en presentaciones.
- tableau-colorblind10 — Paleta accesible para daltónicos.

## Primer gráfico guiado
Graficamos una serie corta para comprobar que todo está listo y usamos la API orientada a objetos (`fig, ax = plt.subplots()`), que mantendremos en las demás lecciones.


````python
# Datos de ejemplo con nombres en español
meses = ["Ene", "Feb", "Mar", "Abr", "May", "Jun"]
ahorros = [100, 50, 300, 10, 150, 500]

# Crear figura y ejes y dibujar la serie
fig, ax = plt.subplots()
ax.plot(meses, ahorros, marker="o", label="Ahorros mensuales")      


# Etiquetas y leyenda
ax.set_title("Ahorros 1er Semestre 2025")  # Título del gráfico
ax.set_xlabel("Meses")                           # Etiqueta del eje X
ax.set_ylabel("Ahorros en USD")                  # Etiqueta del eje Y
ax.legend(['Ahorros Mensuales'])                                      # Muestra la leyenda   
````
